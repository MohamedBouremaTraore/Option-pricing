##### Preprocess options dataset

```python
optionsData['optionType'] = optionsData['source_file'].str.contains('call').replace({True:'call', False:'put'})
```

```python
optionsData['date'] = optionsData['lastTradeDate']
```

```python
optionsData['date'] = pd.to_datetime(optionsData['date']).dt.date
optionsData['date'] = pd.to_datetime(optionsData['date'])
```

```python
optionsData['underlyingTicker'] = optionsData['source_file'].str.split('_').str[0]
```

```python
optionsData['expiration'] = pd.to_datetime(optionsData['expiration'])
```

```python
start_date = pd.to_datetime('2025-09-12')
last_date = pd.to_datetime('2025-12-12')

optionsData = optionsData[optionsData['date'] >= start_date]
optionsData = optionsData[optionsData['date'] <= last_date]
```

```python
optionsData['timeToMaturity'] = (optionsData['expiration'] - optionsData['date']).dt.days/365.25
optionsData = optionsData[optionsData['timeToMaturity'] != 0]
```

##### Handle Put-Call Ratio (PCR) data

```python
df = optionsData
```

```python
pivot = df.pivot_table(index='date', columns='optionType', values='volume', aggfunc='sum').reset_index()

pivot['PCR'] = pivot.apply(lambda row: row['put']/row['call'] if row['call'] != 0 else None, axis=1)

pivot[['date','call','put','PCR']]
```

##### Handle Google Search (GS) Trend data

```python
gs_data_finance = pd.read_csv(gs_folder + "Finance.csv")
gs_data_tech = pd.read_csv(gs_folder + "Technology.csv")
gs_data_auto = pd.read_csv(gs_folder + "Automobile.csv")
gs_data_health = pd.read_csv(gs_folder + "Healthcare.csv")
```

```python
gs_data_finance['date'] = pd.to_datetime(gs_data_finance['date'])
gs_data_tech['date'] = pd.to_datetime(gs_data_tech['date'])
gs_data_auto['date'] = pd.to_datetime(gs_data_auto['date'])
gs_data_health['date'] = pd.to_datetime(gs_data_health['date'])
```

```python
optionsData = optionsData.merge(gs_data_finance, on='date', how='left')
optionsData = optionsData.merge(gs_data_tech, on='date', how='left')
optionsData = optionsData.merge(gs_data_auto, on='date', how='left')
optionsData = optionsData.merge(gs_data_health, on='date', how='left')
```

```python
def get_google_search_trends(row):
    ticker = row['underlyingTicker']
    return row["GS_" + ticker]

optionsData['GS'] = optionsData.apply(get_google_search_trends, axis=1)
```

##### Extract Underlings historical data, VIX, and Interest Rate historical datas

```python
indices_and_equities_underlyings_ticker = TechnologyUnderlying + FinanceUnderlying + HealthUnderlying + AutomobileUnderlying + ["^VIX", "^IRX"]
indices_and_equities_values = {}

for ticker in indices_and_equities_underlyings_ticker:
      data = yf.download(ticker, start=start_date, end=last_date)
      indices_and_equities_values[ticker] = data["Close"]
```

```python
for ticker in indices_and_equities_values:
  temp = indices_and_equities_values[ticker]
  temp = temp.reset_index()
  temp.columns=['date', ticker]
  optionsData = optionsData.merge(temp, on='date', how='left')
```

##### Save dataset

```python
from sklearn.model_selection import train_test_split

list_underlyings = [
    "AAPL",
    "AMZN",
    "GOOGL",
    "META",
    "MSFT",
    "BAC",
    "C",
    "GS",
    "JPM",
    "WFC",
    "ABBV",
    "JNJ",
    "MRK",
    "PFE",
    "UNH",
    "F",
    "GM",
    "LCID",
    "RIVN",
    "TSLA"
]
```

```python
for ticker in list_underlyings:
      columns_to_keep = [
      "date",
      "lastTradeDate",
      "timeToMaturity",
      "underlyingLastPrice",
      "riskFreeRate",
      "lastPrice",
      "impliedVolatility",
      "strike",
      "PCR",
      "VIX",
      "GS_" + ticker,
      "SC_" + ticker
      ]
      data = optionsData[optionsData["underlyingTicker"] == ticker]
      data = data[[col for col in columns_to_keep if col in data.columns]]
      
      train_data, test_data = train_test_split(data, test_size=0.2, random_state=42 )
      train_data.to_csv("./prepocessed options datas/" + ticker + '_train.csv')
      test_data.to_csv("./prepocessed options datas/"  + ticker + '_test.csv')
```
